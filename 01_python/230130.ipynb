# <b>객체 지향 프로그래밍

## <b>Ⅰ. 객체 지향 프로그래밍이란?

### <b>절차 지향 프로그래밍</b>

프로그램 전체가 유기적인 흐름으로 연결  
- 기능 중심의 프로그램  
- 순서가 정해져 있으므로 실행이 빠름  

소프트웨어 위기
- 하드웨어가 발전함에 따라 소프트웨어도 점점 커지고 복잡한 설계 요구
- 하드웨어의 발전 속도를 소프트 웨어의 발전 속도가 따라가지 못 함

절차 지향 방법론의 생산성이 낮음
- 절차 대신 데이터 중심으로 생각
- 객체들을 조립하는 방식으로 개발해보자 >> OOP




### <b>객체 지향 프로그래밍이란?</b>
Object-Oriented Programming(OOP)
- 컴퓨터 프로그램을 명령어 목록으로 보는 시각을 벗어나 여러 개의 독립된 객체로 봄
- 각각의 객체는 메시지를 주고 받고, 데이터를 처리할 수 있음.  

데이터와 기능(메서드) 분리, 추상화된 구조(인터페이스)

### <b>OOP의 장단점</b>
장점
- 잘 만들어놓으면 계속해서 재사용 가능
- 그 자체로 데이터와 행동이 정의됨(독립적) == 내부 구조를 몰라도 다른 객체와 조립,개발 가능
- 객체 단위로 모듈화시켜 개발할 수 있어 많은 인원이 참여하는 대규모 소프트웨어 개발 가능
- 개발 용이성, 유지 보수 편의성, 신뢰성을 바탕으로 생산성이 대폭 증가  

단점
- 설계시 많은 노력과 시간이 필요
- 실행 속도가 상대적으로 느림

## <b>Ⅱ. 객체 지향 프로그래밍이란?

### <b>객체</b>

클래스에서 정의한 것을 토대로 메모리(실제 저장공간)에 할당된 것  
프로그램에서 사용되는 데이터 또는 식별자의 의해 참조되는 공간을 의미  
변수, 자료 구조, 함수 또는 메서드가 될 수 있다.  

속성과 행동으로 구성된 모든 것  

파이썬은 모든 것이 객체(object)

### <b>클래스,객체,인스턴스</b>
클래스로 만든 객체를 인스턴스라고도 함  

클래스(가수)와 객체(실제 사례)
- 이찬혁은 객체다 == 이찬혁은 가수(class)의 인스턴스다.  

하나의 객체는 특정 타입의 인스턴스이다.
- 클래스를 만든다 == 타입을 만든다

### <b>객체의 특징</b>
타입: 어떤 연산자와 조작이 가능한가?  
속성: 어떤 상태(데이터)를 가지는가?  
조작법: 어떤 행위를 할 수 있는가?  

## <b>Ⅲ. OOP 문법

### <b>기본 문법</b>
클래스 정의
- `class Myclass:`  

인스턴스 생성
- `my_instance = Myclass()`  

메서드 호출
- `my_instance.my_method()`  

속성 접근
- `my_instance.my_attribute`


### <b>클래스와 인스턴스</b>
객체의 설계도(클래스)를 가지고, 객체(인스턴스)를 생성
- 클래스: 객체들의 분류/ 설계도(class)
- 인스턴스: 하나하나의 실체/ 예(instance)

### <b>객체 비교하기</b>
==
- 동등한
- 변수가 참조하는 객체가 동등한(내용이 같은) 경우 True
- 두 객체가 같아 보이지만 실제로 동일한 대상을 가리키고 있다고 확인 X  

is
- 동일한
- 두 변수가 동일한 객체를 가리키는 경우 True

### <b>속성</b>
클래스 변수: 클래스 안에 바로 정의  
인스턴스 변수: `self.변수명`

### <b>인스턴스와 클래스 간의 이름 공간</b>
클래스를 정의하면, 클래스와 해당하는 이름 공간 생성  

인스턴스를 만들면, 인스턴스 객체가 생성되고 이름 공간 생성  

인스턴스에서 특정 속성에 접근하면, 인스턴스-클래스 순으로 탐색  

### <b>인스턴스 변수</b>
인스턴스 변수란?
- 인스턴스가 개인적으로 가지고 있는 속성(attribute)
- 각 인스턴스들의 고유한 변수  

생성자 메서드(`__init__`)에서 `self.<name>`으로 정의  

인스턴스가 생성된 이후 `<instance>.<name>`으로 접근 및 할당

### <b>클래스 변수</b>
클래스 변수란?
- 한 클래스의 모든 인스턴스가 공유하는 값을 의미
- 같은 클래스의 인스턴스들은 같은 값을 갖게 됨
- 예) 특정 사이트의 User수 등은 클래스 변수를 사용해야 함
- 클래스 선언 내부에서 정의
- `<classname>.<name>`으로 접근 및 할당  

클래스 변수를 변경
- 항상 `<classname>.<name>`으로 변경

## <b>Ⅳ. OOP 메서드

### <b>메서드</b>
특정 데이터 타입/클래스의 객체에 공통적으로 적용 가능한 행위(함수)
- 인스턴스 메서드: 인스턴스 변수를 사용하는 함수
- 클래스 메서드: 클래스 변수를 사용하는 함수
- 정적 메서드: 두 변수 모두 사용 X

### <b>인스턴스 메서드</b>
인스턴스 변수를 사용하거나, 인스턴스 변수에 값을 설정하는 메서드  

클래스 내부에 정의되는 메서드의 기본  

호출 시, 첫번째 인자로 인스턴스 자기 자신(self)이 자동으로 전달됨


### <b>self</b>
인스턴스 자기자신  
파이썬에서 인스턴스 메서드는 호출 시 첫번째 인자로 인스턴스 자신이 전달되게 설계
- 매개 변수 이름으로 self를 첫 번째 인자로 정의
- 다른 단어로 써도 작동하지만, 파이썬의 암묵적인 규칙

### <b>매직 메서드</b>
더블 언더스코어(__)가 있는 메서드는 특수한 동적을 위해 만들어짐  
스페셜 메서드 혹은 매직 메서드  
특정 상황에 자동으로 불리는 메서드  

객체 특수 조작 행위를 지정
- `__str__` : 객체를 문자열로 어떻게 표현할지 지정  
print함수 등에서 객체를 출력하면 자동으로 호출
- `__gt__` : 부등호 연산자(>,greater than)
- `__init__`(생성자 메서드) : 인스턴스 객체가 생성될 때 자동으로 호출되는 메서드  
인스턴스 변수들의 초기값을 설정

### <b>클래스 메서드</b>
- 클래스가 사용할 메서드

- `@classmethod` 데코레이터를 사용하여 정의

- 호출 시, 첫번째 인자로 클래스(cls)가 전달됨



#### <b>데코레이터</b>
함수를 어떤 함수로 꾸며서 새로운 기능을 부여  

@데코레이터(함수명) 형태로 함수 위에 작성  

순서대로 적용 되기 때문에 작성 순서가 중요
